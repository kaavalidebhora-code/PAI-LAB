```python

goal_A = "Clean"
goal_B = "Clean"

def dfs(state, path=[], visited=set()):
    location, A, B = state


    if A == goal_A and B == goal_B:
        return path, state

    if state in visited:
        return None

    visited.add(state)


    neighbors = []


    if location == "A" and A == "Dirty":
        neighbors.append(("Suck", ("A", "Clean", B)))
    if location == "B" and B == "Dirty":
        neighbors.append(("Suck", ("B", A, "Clean")))


    if location == "A":
        neighbors.append(("Right", ("B", A, B)))

    if location == "B":
        neighbors.append(("Left", ("A", A, B)))


    for action, next_state in neighbors:
        result = dfs(next_state, path + [(action, next_state)], visited)
        if result:
            return result

    return None


states = [
    ("A","Dirty","Dirty"),
    ("A","Dirty","Clean"),
    ("A","Clean","Dirty"),
    ("A","Clean","Clean"),
    ("B","Dirty","Dirty"),
    ("B","Dirty","Clean"),
    ("B","Clean","Dirty"),
    ("B","Clean","Clean")
]

for start in states:
    print("\nInitial State:", start)
    result = dfs(start)

    if result:
        path, goal = result
        step = 1
        for action, state in path:
            print("Step", step, "Action:", action, "->", state)
            step += 1
        print("Goal reached:", goal)
    else:
        print("No solution found")
```
