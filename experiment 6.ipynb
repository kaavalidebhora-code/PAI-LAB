```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Rain': ['No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'TrafficJam': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No'],
    'ArriveLate': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
})

# Define Bayesian Network
model = DiscreteBayesianNetwork([
    ('Rain', 'TrafficJam'),
    ('TrafficJam', 'ArriveLate')
])

# Fit model (VERY IMPORTANT)
model.fit(data, estimator=MaximumLikelihoodEstimator)

# Print CPDs
print(model.get_cpds())

# Inference
inference = VariableElimination(model)

query_result = inference.query(
    variables=['ArriveLate'],
    evidence={'Rain': 'Yes'}
)

print(query_result)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd
data = pd.DataFrame({
    'Age': ['Over 50', 'Under 50', 'Under 50', 'Over 50', 'Under 50'],
    'Smoking': ['Yes', 'Yes', 'No', 'Yes', 'No'],
    'HeartDisease': ['Yes', 'No', 'No', 'Yes', 'No']
})
model = DiscreteBayesianNetwork([
    ('Age', 'HeartDisease'),
    ('Smoking', 'HeartDisease')
])
model.fit(data, estimator=MaximumLikelihoodEstimator)


# Print CPDs (Conditional Probability Distributions)
for cpd in model.get_cpds():
    print(cpd)

# Perform Inference
inference = VariableElimination(model)

# Query: What is the probability of having Heart Disease given Age = Over 50 and Smoking = Yes?
query_result = inference.query(variables=['HeartDisease'], evidence={'Age': 'Over 50', 'Smoking': 'Yes'})

print("\nInference Result (Probability of Heart Disease):")
print(query_result)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

data = pd.DataFrame({
    'ProductQuality': ['High', 'Low', 'High', 'Low', 'High'],
    'CustomerService': ['Good', 'Bad', 'Good', 'Bad', 'Good'],
    'ReviewRating': ['Positive', 'Negative', 'Positive', 'Negative', 'Positive']
})


model = DiscreteBayesianNetwork([
    ('ProductQuality', 'ReviewRating'),
    ('CustomerService', 'ReviewRating')
])


model.fit(data, estimator=MaximumLikelihoodEstimator)


for cpd in model.get_cpds():
    print(cpd)


inference = VariableElimination(model)


query_result = inference.query(variables=['ReviewRating'], evidence={'ProductQuality': 'High', 'CustomerService': 'Good'})

print("\nInference Result (Probability of Positive Review):")
print(query_result)
```

```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

# Create a simple Bayesian Network model
model = BayesianNetwork([('Credit Score', 'Default Status'),
                         ('Income', 'Default Status')])

# Define the CPDs (conditional probability distributions)
cpd_credit_score = TabularCPD(variable='Credit Score', variable_card=2,
                              values=[[0.7], [0.3]])  # 70% good credit, 30% bad credit

cpd_income = TabularCPD(variable='Income', variable_card=2,
                        values=[[0.6], [0.4]])  # 60% high income, 40% low income

# P(Default Status | Credit Score, Income)
cpd_default_status = TabularCPD(variable='Default Status', variable_card=2,
                                values=[[0.9, 0.8],  # Non-default (0)
                                        [0.1, 0.2]],  # Default (1)
                                evidence=['Credit Score', 'Income'],
                                evidence_card=[2, 2])

# Add CPDs to the model
model.add_cpds(cpd_credit_score, cpd_income, cpd_default_status)

# Check if the model is valid
assert model.check_model()

# Perform inference
inference = VariableElimination(model)

# Query the model: What is the probability of Default Status given Credit Score=1 (good) and Income=1 (high)?
query = inference.query(variables=['Default Status'], evidence={'Credit Score': 1, 'Income': 1})

print(query)
```
