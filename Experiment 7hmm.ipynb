```python
import numpy as np

# -----------------------
# 1. States & Observations
# -----------------------
states = ['Rainy', 'Sunny']
observations = ['walk', 'shop', 'clean']

state_idx = {s:i for i,s in enumerate(states)}
obs_idx = {o:i for i,o in enumerate(observations)}

# -----------------------
# 2. Initial Probabilities (π)
# -----------------------
pi = np.array([0.6, 0.4])

# -----------------------
# 3. Transition Matrix (A)
# -----------------------
A = np.array([
    [0.7, 0.3],
    [0.4, 0.6]
])

# -----------------------
# 4. Emission Matrix (B)
# -----------------------
B = np.array([
    [0.1, 0.4, 0.5],
    [0.6, 0.3, 0.1]
])

# -----------------------
# 5. Observation Sequence
# -----------------------
O = [obs_idx['walk'], obs_idx['shop'], obs_idx['clean']]

T = len(O)
N = len(states)

# =======================
# FORWARD ALGORITHM
# =======================
alpha = np.zeros((T, N))

# Initialization
alpha[0] = pi * B[:, O[0]]

# Recursion
for t in range(1, T):
    for j in range(N):
        alpha[t, j] = np.sum(alpha[t-1] * A[:, j]) * B[j, O[t]]

forward_prob = np.sum(alpha[T-1])

# =======================
# BACKWARD ALGORITHM
# =======================
beta = np.zeros((T, N))

# Initialization
beta[T-1] = 1

# Recursion
for t in range(T-2, -1, -1):
    for i in range(N):
        beta[t, i] = np.sum(A[i] * B[:, O[t+1]] * beta[t+1])

backward_prob = np.sum(pi * B[:, O[0]] * beta[0])

# =======================
# OUTPUT
# =======================
print("Forward Matrix (alpha):")
print(alpha)

print("\nBackward Matrix (beta):")
print(beta)

print("\nForward Probability:", forward_prob)
print("Backward Probability:", backward_prob)
```

```python
import numpy as np

# States and Observations
states = ['Normal', 'Fraud']
observations = ['Low', 'Medium', 'High']

# Index mapping
state_idx = {s:i for i,s in enumerate(states)}
obs_idx = {o:i for i,o in enumerate(observations)}

# Initial probabilities
pi = np.array([0.8, 0.2])

# Transition matrix
A = np.array([[0.7, 0.3],
              [0.4, 0.6]])

# Emission matrix
B = np.array([[0.6, 0.3, 0.1],
              [0.1, 0.3, 0.6]])

# Observation sequence: Low → Medium → High
O = [obs_idx['Low'], obs_idx['Medium'], obs_idx['High']]

T = len(O)
N = len(states)

# -----------------------
# Forward Algorithm
# -----------------------
alpha = np.zeros((T, N))

# Step 1: Initialization
alpha[0] = pi * B[:, O[0]]

# Step 2: Recursion
for t in range(1, T):
    for j in range(N):
        alpha[t, j] = np.sum(alpha[t-1] * A[:, j]) * B[j, O[t]]

# Step 3: Termination
forward_prob = np.sum(alpha[T-1])

print("Forward Probability:", forward_prob)

# -----------------------
# Backward Algorithm
# -----------------------
beta = np.zeros((T, N))

# Step 1: Initialization
beta[T-1] = 1

# Step 2: Recursion
for t in range(T-2, -1, -1):
    for i in range(N):
        beta[t, i] = np.sum(A[i] * B[:, O[t+1]] * beta[t+1])

# Step 3: Termination
backward_prob = np.sum(pi * B[:, O[0]] * beta[0])

print("Backward Probability:", backward_prob)
```
