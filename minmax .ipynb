```python
import math

def minimax(stones, isMax, depth=0):
    # Base case: no stones left
    if stones == 0:
        return -1 if isMax else 1  # if MAX cannot move, MIN wins

    moves = [1, 2]  # possible stones to remove
    moves = [m for m in moves if m <= stones]

    indent = "  " * depth  # for printing tree

    if isMax:
        best = -math.inf
        for move in moves:
            score = minimax(stones - move, False, depth + 1)
            best = max(best, score)
        print(f"{indent}MAX stones={stones} -> best score={best}")
        return best
    else:
        best = math.inf
        for move in moves:
            score = minimax(stones - move, True, depth + 1)
            best = min(best, score)
        print(f"{indent}MIN stones={stones} -> best score={best}")
        return best

def best_move(stones):
    best_score = -math.inf
    move_to_take = 1
    for move in [1,2]:
        if move <= stones:
            score = minimax(stones - move, False)
            if score > best_score:
                best_score = score
                move_to_take = move
    return move_to_take

# ---- Game Loop ----
stones = int(input("Enter initial number of stones (5-10 recommended): "))
print(f"Game start! There are {stones} stones.\n")

while stones > 0:
    # Player move
    player_move = int(input("Your turn! Remove 1 or 2 stones: "))
    if player_move not in [1,2] or player_move > stones:
        print("Invalid move! Try again.")
        continue
    stones -= player_move
    print(f"Stones remaining: {stones}\n")
    if stones == 0:
        print("You took the last stone. You win! 🎉")
        break

    # AI move
    print("AI is thinking...\n")
    ai_move = best_move(stones)
    print(f"AI removes {ai_move} stones.")
    stones -= ai_move
    print(f"Stones remaining: {stones}\n")
    if stones == 0:
        print("AI took the last stone. AI wins! 🤖")
        break
```
