```python
import math

def alphabeta(stones, isMax, alpha, beta, depth=0):
    # Base case: no stones left
    if stones == 0:
        return -1 if isMax else 1  # current player loses

    moves = [1, 2]
    moves = [m for m in moves if m <= stones]
    indent = "  " * depth  # for visualizing the tree

    if isMax:
        maxEval = -math.inf
        for move in moves:
            eval = alphabeta(stones - move, False, alpha, beta, depth + 1)
            maxEval = max(maxEval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:  # prune
                print(f"{indent}MAX pruning at stones={stones}, move={move}")
                break
        print(f"{indent}MAX stones={stones} -> best score={maxEval}")
        return maxEval
    else:
        minEval = math.inf
        for move in moves:
            eval = alphabeta(stones - move, True, alpha, beta, depth + 1)
            minEval = min(minEval, eval)
            beta = min(beta, eval)
            if beta <= alpha:  # prune
                print(f"{indent}MIN pruning at stones={stones}, move={move}")
                break
        print(f"{indent}MIN stones={stones} -> best score={minEval}")
        return minEval

def best_move(stones):
    best_score = -math.inf
    move_to_take = 1
    for move in [1, 2]:
        if move <= stones:
            score = alphabeta(stones - move, False, -math.inf, math.inf)
            if score > best_score:
                best_score = score
                move_to_take = move
    return move_to_take

# ---- Game Loop ----
stones = int(input("Enter initial number of stones (5-15 recommended): "))
print(f"Game start! There are {stones} stones.\n")

while stones > 0:
    # Player move
    player_move = int(input("Your turn! Remove 1 or 2 stones: "))
    if player_move not in [1, 2] or player_move > stones:
        print("Invalid move! Try again.")
        continue
    stones -= player_move
    print(f"Stones remaining: {stones}\n")
    if stones == 0:
        print("You took the last stone. You win! 🎉")
        break

    # AI move
    print("AI is thinking...\n")
    ai_move = best_move(stones)
    print(f"AI removes {ai_move} stones.")
    stones -= ai_move
    print(f"Stones remaining: {stones}\n")
    if stones == 0:
        print("AI took the last stone. AI wins! 🤖")
        break
```
