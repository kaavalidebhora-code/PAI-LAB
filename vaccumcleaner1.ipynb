```python
import heapq

# Goal states
goal_A = "Clean"
goal_B = "Clean"

# Heuristic: number of dirty rooms
def heuristic(state):
    _, A, B = state
    h = 0
    if A == "Dirty":
        h += 1
    if B == "Dirty":
        h += 1
    return h

# Generate neighbors
def get_neighbors(state):
    location, A, B = state
    neighbors = []

    # Suck
    if location == "A" and A == "Dirty":
        neighbors.append(("Suck", ("A", "Clean", B)))
    if location == "B" and B == "Dirty":
        neighbors.append(("Suck", ("B", A, "Clean")))

    # Move Right
    if location == "A":
        neighbors.append(("Right", ("B", A, B)))

    # Move Left
    if location == "B":
        neighbors.append(("Left", ("A", A, B)))

    return neighbors

# A* search
def a_star(start):
    # Priority queue: (f = g+h, g, state, path)
    open_list = []
    heapq.heappush(open_list, (heuristic(start), 0, start, []))
    visited = set()

    while open_list:
        f, g, state, path = heapq.heappop(open_list)
        if state in visited:
            continue
        visited.add(state)

        location, A, B = state

        # Goal test
        if A == goal_A and B == goal_B:
            return path, state

        for action, next_state in get_neighbors(state):
            g_next = g + 1
            h_next = heuristic(next_state)
            f_next = g_next + h_next
            heapq.heappush(open_list, (f_next, g_next, next_state, path + [(action, next_state)]))

    return None

# Test all starting states
states = [
    ("A","Dirty","Dirty"),
    ("A","Dirty","Clean"),
    ("A","Clean","Dirty"),
    ("A","Clean","Clean"),
    ("B","Dirty","Dirty"),
    ("B","Dirty","Clean"),
    ("B","Clean","Dirty"),
    ("B","Clean","Clean")
]

for start in states:
    print("\nInitial State:", start)
    result = a_star(start)

    if result:
        path, goal = result
        step = 1
        for action, state in path:
            print(f"Step {step} Action: {action} -> {state}")
            step += 1
        print("Goal reached:", goal)
    else:
        print("No solution found")
```
