```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Rain': ['No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'TrafficJam': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No'],
    'ArriveLate': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
})

# Define Bayesian Network
model = DiscreteBayesianNetwork([
    ('Rain', 'TrafficJam'),
    ('TrafficJam', 'ArriveLate')
])

# Fit model (VERY IMPORTANT)
model.fit(data, estimator=MaximumLikelihoodEstimator)

# Print CPDs
print(model.get_cpds())

# Inference
inference = VariableElimination(model)

query_result = inference.query(
    variables=['ArriveLate'],
    evidence={'Rain': 'Yes'}
)

print(query_result)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd
data = pd.DataFrame({
    'Age': ['Over 50', 'Under 50', 'Under 50', 'Over 50', 'Under 50'],
    'Smoking': ['Yes', 'Yes', 'No', 'Yes', 'No'],
    'HeartDisease': ['Yes', 'No', 'No', 'Yes', 'No']
})
model = DiscreteBayesianNetwork([
    ('Age', 'HeartDisease'),
    ('Smoking', 'HeartDisease')
])
model.fit(data, estimator=MaximumLikelihoodEstimator)


# Print CPDs (Conditional Probability Distributions)
for cpd in model.get_cpds():
    print(cpd)

# Perform Inference
inference = VariableElimination(model)

# Query: What is the probability of having Heart Disease given Age = Over 50 and Smoking = Yes?
query_result = inference.query(variables=['HeartDisease'], evidence={'Age': 'Over 50', 'Smoking': 'Yes'})

print("\nInference Result (Probability of Heart Disease):")
print(query_result)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

# Sample Data
data = pd.DataFrame({
    'ProductQuality': ['High', 'Low', 'High', 'Low', 'High'],
    'CustomerService': ['Good', 'Bad', 'Good', 'Bad', 'Good'],
    'ReviewRating': ['Positive', 'Negative', 'Positive', 'Negative', 'Positive']
})

# Define Bayesian Network Structure
model = DiscreteBayesianNetwork([
    ('ProductQuality', 'ReviewRating'),
    ('CustomerService', 'ReviewRating')
])

# Fit the model
model.fit(data, estimator=MaximumLikelihoodEstimator)

# Print CPDs (Conditional Probability Distributions)
for cpd in model.get_cpds():
    print(cpd)

# Perform Inference
inference = VariableElimination(model)

# Query: What is the probability of Positive Review given High Product Quality and Good Customer Service?
query_result = inference.query(variables=['ReviewRating'], evidence={'ProductQuality': 'High', 'CustomerService': 'Good'})

print("\nInference Result (Probability of Positive Review):")
print(query_result)
```
