```python
import math

def minimax(curDepth, nodeIndex, maxTurn, scores, targetDepth):

    # Base case
    if curDepth == targetDepth:
        return scores[nodeIndex]

    if maxTurn:
        return max(
            minimax(curDepth + 1, nodeIndex * 2, False, scores, targetDepth),
            minimax(curDepth + 1, nodeIndex * 2 + 1, False, scores, targetDepth)
        )
    else:
        return min(
            minimax(curDepth + 1, nodeIndex * 2, True, scores, targetDepth),
            minimax(curDepth + 1, nodeIndex * 2 + 1, True, scores, targetDepth)
        )


# 🔹 Take number of leaf nodes
n = int(input("Enter number of leaf nodes (must be power of 2): "))

# 🔹 Check if power of 2
if (n & (n - 1)) != 0:
    print("Number must be a power of 2!")
    exit()

# 🔹 Take leaf node values
scores = list(map(int, input(f"Enter {n} leaf values separated by space: ").split()))

if len(scores) != n:
    print("Error: Number of values entered does not match n")
    exit()

# 🔹 Calculate depth
treeDepth = int(math.log2(n))

# 🔹 Compute result
optimal_value = minimax(0, 0, True, scores, treeDepth)

print("The optimal value is:", optimal_value)
```
